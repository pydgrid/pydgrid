## Solar irradiation model

Following model is from https://dspace.lboro.ac.uk/dspace-jspui/handle/2134/7773

```python
import numpy as np
import matplotlib.pyplot as plt
from pydgrid import solar
import pandas as pd
```

```python
%matplotlib notebook
```

### Single value of total radiation on panel

```python
pv = solar.pv_gen()
pv.Slope_of_panel = 10
pv.Latitude =  37.3828300 # Sevilla
pv.Longitude = -5.9731700 # Sevilla

day_of_the_year = 180.0  # june 30
hour_of_day = 10.0       # 10:00 am
minute_of_hour = 0      

q = pv.radiation_on_panel_eval(day_of_the_year,hour_of_day,minute_of_hour)
print('Total radiation on panel: {:0.2f} W/m²'.format(q))
```

### Single value of total radiation on panel

```python
drange = pd.date_range(start='2018-01-01',end='2018-12-31', freq='h')
radiations_on_panel_clear = pv.radiations_clear(drange)
  
fig, ax = plt.subplots()
ax.plot(drange[1:],pv.radiation_on_panel_clear_sky)
ax.set_xlabel('Date')
ax.set_ylabel('Radiation on panel (W/m$^2$)')
```
