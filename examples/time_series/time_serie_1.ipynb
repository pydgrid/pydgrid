```python
import numpy as np
from pydgrid import grid
from pydgrid.pf import time_serie
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
sys1 = grid()
sys1.read('cigre_europe_residential.json')  # Load data
sys1.read_loads_shapes('load_shapes_list.json')

t_ini = 0.0
t_ini = 0.0
t_end = 24*60*60
Dt = 10

%timeit time_serie(t_ini,t_end,Dt,sys1.params_pf,sys1.params_lshapes)
```

```python
T,V_nodes,I_nodes,Iters = time_serie(t_ini,t_end,Dt,sys1.params_pf,sys1.params_lshapes,interp=1)
```

```python
fig, axes = plt.subplots(nrows=2, figsize=(7,7.0))



axes[0].plot(T/60,np.abs(I_nodes))
axes[1].plot(T/60,np.abs(V_nodes))

axes[0].grid(True)
axes[1].grid(True)

axes[1].set_ylim((150,250.0))

axes[0].set_ylabel('Currents (A)')
axes[1].set_ylabel('Voltages Ph-N (V)')

axes[1].set_xlabel('Time (min)')
```
